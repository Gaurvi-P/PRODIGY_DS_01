## Dropping Unnecessary columns

```python
df.drop(['Series Name','Country Name'],axis=1,inplace=True)
```

## Extraction of top 10 countries with respect to total population

```python
#filter data for total population
total_population_data=df[df['Series Code']=='SP.POP.TOTL']

#Sort data based on the total population for 2022
total_population_sorted=total_population_data.sort_values(by="2022", ascending=False)

#Get the top ten countries with the highest total population for 2022
total_top_ten_countries=total_population_sorted.head(10)

print("Top ten countries of total population \n")
print(total_top_ten_countries[['Country Code']])
```

# Bar Plot

## Top ten countriesof total population in year 2022 and 2012

```python
#create a bar plot
plt.figure(figsize=(15,6)) #figure of given data
plt.subplot(2,2,1) #defining dimension of the chart
sns.barplot(x="2022", y="Country Code", data=total_top_ten_countries,palette="coolwarm")
plt.title("Top 10 countries of total population(2022)",fontsize=12)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country Code",fontsize=10)
plt.show()

#create a bar plot
plt.figure(figsize=(15,6)) #figure of given data
plt.subplot(2,2,1) #defining dimension of the chart
sns.barplot(x="2012", y="Country Code", data=total_top_ten_countries,palette="coolwarm")
plt.title("Top 10 countries of total population(2012)",fontsize=12)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country Code",fontsize=10)
plt.show()
```

## extraction of bottom 10 countries

```python
#Sort data based on the total population for 2022
total_population_sorted1=total_population_data.sort_values(by="2022", ascending=True)

#Get the bottom ten countries with the highest total population for 2022
total_bottom_ten_countries=total_population_sorted1.head(10)

print("Bottom ten countries of total population \n")
print(total_bottom_ten_countries[['Country Code']])

#create a bar plot
plt.figure(figsize=(15,6)) #figure of given data
plt.subplot(2,2,1) #defining dimension of the chart
sns.barplot(x="2022", y="Country Code", data=total_bottom_ten_countries,palette="viridis")
plt.title("Bottom 10 countries of total population(2022)",fontsize=12)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country Code",fontsize=10)
plt.show()

#create a bar plot
plt.figure(figsize=(15,6)) #figure of given data
plt.subplot(2,2,1) #defining dimension of the chart
sns.barplot(x="2012", y="Country Code", data=total_bottom_ten_countries,palette="viridis")
plt.title("Bottom 10 countries of total population(2012)",fontsize=12)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country Code",fontsize=10)
plt.show()
```

## Extraction of top 10 countries with highest male and female population

```python
#filter data for male population
male_population_data=df[df['Series Code']=='SP.POP.TOTL.MA.IN']

#Sort data based on the male population for 2022
male_population_sorted=male_population_data.sort_values(by="2022", ascending=False)

#Get the top ten countries with the highest male population for 2022
male_top_ten_countries=male_population_sorted.head(10)

print("Top ten countries with highest male population \n")
print(male_top_ten_countries[['Country Code']])

#filter data for female population
female_population_data=df[df['Series Code']=='SP.POP.TOTL.FE.IN']

#Sort data based on the female population for 2022
female_population_sorted=female_population_data.sort_values(by="2022", ascending=False)

#Get the top ten countries with the highest female population for 2022
female_top_ten_countries=female_population_sorted.head(10)

print("Top ten countries with highest female population \n")
print(female_top_ten_countries[['Country Code']])
```

## Top ten countries with highest male and female population 2022

```python

#create a bar plot
plt.figure(figsize=(15,6)) #figure of given data
plt.subplot(2,2,1) #defining dimension of the chart
sns.barplot(x="2022", y="Country Code", data=male_top_ten_countries,palette="coolwarm")
plt.title("Top 10 countries with highest male population(2022)",fontsize=12)
plt.xlabel("Male population",fontsize=10)
plt.ylabel("Country Code",fontsize=10)
plt.show()

#create a bar plot
plt.figure(figsize=(15,6)) #figure of given data
plt.subplot(2,2,1) #defining dimension of the chart
sns.barplot(x="2012", y="Country Code", data=female_top_ten_countries,palette="coolwarm")
plt.title("top 10 countries with highest female population(2012)",fontsize=12)
plt.xlabel("Female population",fontsize=10)
plt.ylabel("Country Code",fontsize=10)
plt.show()
```

# Stacked Bar Plot

## Top 10 countries with male and female population

```python
#merge male and female data on 'Country Code'
merge_data=pd.merge(male_population_data,female_population_data,on="Country Code",suffixes=("_male","_female"))

#merged data
#Calculate the total population for each country (male + female)
merge_data["Total Population"]=merge_data["2022_male"]+merge_data["2022_female"]

#Checking top 5 data
merge_data.head(5)
```

```python
#sort data based on total population in descending order
sorted_data=merge_data.sort_values(by="Total Population", ascending=False)
```

## Select the top 10 countries with the highest total population

```python
top_ten_countries=sorted_data.head(10)
```

```python
#create a stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_male",data=top_ten_countries,color="red",label="male Population")
sns.barplot(x="Country Code",y="2022_female",data=top_ten_countries,bottom=top_ten_countries["2022_male"],color="yellow",label="female Population")
plt.xlabel("Country Code")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

## Bottom 10 countries with male and female population(2022)

```python
#select top 10 countries wuth the highest total population
bottom_ten_countries=sorted_data.tail(10)

#create a stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_male",data=bottom_ten_countries,color="red",label="male Population")
sns.barplot(x="Country Code",y="2022_female",data=bottom_ten_countries,bottom=bottom_ten_countries["2022_male"],color="yellow",label="female Population")
plt.xlabel("Country Code")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```
